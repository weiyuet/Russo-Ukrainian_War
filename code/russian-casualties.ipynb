```python
# Load libraries
import numpy as np
import pandas as pd

import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
# Load data
# Load data Russian casualties
russian_casualties = pd.read_csv('~/Documents/R/russo-ukrainian-war/data/russia_losses_personnel.csv')

# Load data Russian equipment
russian_equipment = pd.read_csv('~/Documents/R/russo-ukrainian-war/data/russia_losses_equipment.csv')
```

```python
# Plot Russian casualties
x = russian_casualties['date']
y0 = russian_casualties['personnel']
y1 = russian_casualties['POW']

fig = go.Figure()

fig.add_trace(go.Scatter(x = x, y = y0,
                         mode = 'lines + markers',
                         name = 'Personnel'))

fig.add_trace(go.Scatter(x = x, y = y1,
                         mode = 'lines + markers',
                         name = 'POW'))

fig.update_layout(legend_orientation = 'h',
                  legend = dict(x = 0, y = 1),
                  title = 'Russian Casualties')

fig.show()
```

```python
# Plot Russian equipment lost - aerial weapons
x = russian_equipment['date']
y0 = russian_equipment['aircraft']
y1 = russian_equipment['helicopter']
y2 = russian_equipment['anti-aircraft warfare']
y3 = russian_equipment['drone']

fig = go.Figure()

fig.add_trace(go.Scatter(x = x, y = y0,
                         mode = 'lines + markers',
                         name = 'Aircraft'))
fig.add_trace(go.Scatter(x = x, y = y1,
                         mode = 'lines + markers',
                         name = 'Helicopter'))
fig.add_trace(go.Scatter(x = x, y = y2,
                         mode = 'lines + markers',
                         name = 'Anti-aircraft warfare'))
fig.add_trace(go.Scatter(x = x, y = y3,
                         mode = 'lines + markers',
                         name = 'Drone'))

fig.update_layout(legend_orientation = 'h',
                  legend = dict(x = 0, y = 1),
                  title = 'Russian Equipment Lost - Aerial Weapons')

fig.show()
```
